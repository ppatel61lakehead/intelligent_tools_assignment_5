```
import math
import numpy as np
#initialize weights
w3 = [0.5,0.5,0.5]
w4 = [0.5,0.5,0.5]
w6 = [0.5,0.5,0.5]
# w3 = [0.2,0.2,0.2]
# w4 = [0.2,0.2,0.2]
# w6 = [0.2,0.2,0.2]
E = 0
n = 0.2
```

```
def sigmoid(x,lamda):
  return 1/(1+math.pow(math.e,-(lamda*x)))
```

```
def forward(x,weight):
  X = np.matrix(x)
  W = np.matrix(weight)
  return np.dot(X,W.T)
```

```
def all(o0,o1,o2,t):
  global w3,w4,w6,n,E,outputs
  E = 0
  o3 = round(sigmoid(forward([o0,o1,o2],w3),1),4)
  o4 = round(sigmoid(forward([o0,o1,o2],w4),1),4)
  o5 = -1
  print("Forward propagation:")
  print("O3: "+str(o3))
  print("O4: "+str(o4))
  hiddenInput = [o3,o4,o5]
  o6 = round(sigmoid(forward(hiddenInput,w6),1),4)
  print("O6: "+str(o6))

  #Error Measure
  print("Output error measure:")

  E = round((0.5*(t-o6)*(t-o6))+E,4)
  print("E: "+str(E))
  delta_6 = round(o6*(1-o6)*(t-o6),4)
  print("delta-6: "+str(delta_6))

  #Backpropogation
  print("\nThird layer weight updates:")
  diffw63 = n*delta_6*o3
  diffw64 = n*delta_6*o4
  diffw65 = n*delta_6*o5
  w6 = [round(w6[0]+diffw63,4),round(w6[1]+diffw64,4),round(w6[2]+diffw65,4)]
  print("delta_W63: "+str(diffw63))
  print("delta_W64: "+str(diffw64))
  print("delta_W65: "+str(diffw65)+"\n")
  print("new_W63: "+str(w6[0]))
  print("new_W64: "+str(w6[1]))
  print("new_W65: "+str(w6[2]))

  print("\nSecond layer error signals:")
  delta_3 = round(o3*(1-o3)*w6[0]*delta_6,4)
  delta_4 = round(o4*(1-o4)*w6[1]*delta_6,4)
  print("delta_3: "+str(delta_3))
  print("delta_4: "+str(delta_4))

  print("\nSecond layer weight updates:")
  diffw30 = n*delta_3*o0
  diffw31 = n*delta_3*o1
  diffw32 = n*delta_3*o2
  w3 = [round(w3[0]+diffw30,4),round(w3[1]+diffw31,4),round(w3[2]+diffw32,4)]
  print("delta_W30: "+str(diffw30))
  print("delta_W31: "+str(diffw31))
  print("delta_W32: "+str(diffw32)+"\n")
  print("new_W30: "+str(w3[0]))
  print("new_W31: "+str(w3[1]))
  print("new_W32: "+str(w3[2]))

  diffw40 = n*delta_4*o0
  diffw41 = n*delta_4*o1
  diffw42 = n*delta_4*o2
  w4 = [round(w4[0]+diffw40,4),round(w4[1]+diffw41,4),round(w4[2]+diffw42,4)]
  print("delta_W40: "+str(diffw40))
  print("delta_W41: "+str(diffw41))
  print("delta_W42: "+str(diffw42)+"\n")
  print("new_W40: "+str(w4[0]))
  print("new_W41: "+str(w4[1]))
  print("new_W42: "+str(w4[2]))
```

```
def epoch(e):
  print("\nEpoch::: "+str(e)+"\n")
  print("\nTaking First Input:")
  print("-------------------")
  all(0.3,0.4,-1,0.88)
  print("\nTaking Second Input:")
  print("-------------------")
  all(0.1,0.6,-1,0.82)
  print("\nTaking Third Input:")
  print("-------------------")
  all(0.9,0.4,-1,0.57)
```

```
epoch(1)
epoch(2)
epoch(3)
epoch(4)
epoch(5)
epoch(6)
epoch(7)
epoch(8)
epoch(9)
epoch(10)
```
